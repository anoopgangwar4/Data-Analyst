```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
df
```

```python
#check for null values
df.isnull().sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# change datatype
df['Amount']=df['Amount'].astype(int)
df['Amount'].dtype
```

```python
df.columns
```

```python
#rename columns
df.rename(columns={'Marital_Status':'shaadi'})
```

# Exploratory Data Analysis
#Gender

```python
# plotting a bar chart for Gender and it's count
num=sns.countplot(x='Gender',data=df)
for bar in num.containers:
   num.bar_label(bar)
```

```python
# plotting a bar chart for gender vs total amount
group=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
num=sns.barplot(x='Gender',y='Amount',data=group)
for i in num.containers:
    num.bar_label(i)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

# Age

```python

num=sns.countplot(data=df,x='Age Group',hue='Gender')
for i in num.containers:
    num.bar_label(i)
```

```python
# Total Amount vs Age Group
total=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
num=sns.barplot(data=total,x='Age Group',y='Amount')
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set({'figure.figsize':(15,5)})
num=sns.barplot(data = sales_state, x = 'State',y= 'Orders')
for i in num.containers:
    num.bar_label(i)
```

```python
# total amount/sales from top 10 states
sales_amt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
num=sns.barplot(data=sales_amt,x='State',y='Amount')
sns.set({'figure.figsize':(15,5)})
plt.xticks(rotation=90)
for i in num.containers:
    num.bar_label(i)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital Status

```python
num=sns.countplot(data=df,x='Marital_Status')
for i in num.containers:
    num.bar_label(i)
```

```python
grp=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
num=sns.barplot(data=grp,x='Marital_Status',y='Amount',hue='Gender')
for i in num.containers:
    num.bar_label(i)
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
occu=sns.countplot(data=df,x='Occupation')
for i in occu.containers:
    occu.bar_label(i)
plt.xticks(rotation=45);
```

```python
occuGrp=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=occuGrp,x='Occupation',y='Amount')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product category

```python
sns.countplot(data=df,x='Product_Category')
plt.xticks(rotation=45);
```

```python
pro=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
pro=sns.barplot(data=pro,x='Product_Category',y='Amount');
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
# top 10 most sold products 
sales_status=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_status,x='Product_ID',y='Orders');
```
